# Segmentation

## Unsupervised segmentation

###  SLIC algorithm 

```python
from skimage import io, segmentation, color
```

```python
from skimage.io import imread
```

```python
from matplotlib import pyplot as plt
```

```python
img=imread('img/animal.png')
```

```python
io.imshow(img) # show the input image
```

```python
image_slic=segmentation.slic(img,n_segments=20)
```

```python
superpixels=color.label2rgb(image_slic, img, kind='avg')
```

```python
io.imshow(superpixels)
```

## contour detection

```python
from skimage import measure, io, color
```

```python
from skimage.io import imread
```

```python
from skimage.color import rgb2gray
```

```python
from skimage.filters import sobel
```

```python
import matplotlib.pyplot as plt
```

```python
img = io.imread('img/contour.png')  #Read an image
```

```python
io.imshow(img)
```

```python
img_gray = rgb2gray(img) #Convert the image to grayscale
```

```python
img_edges = sobel(img_gray) #Find edges in the image
```

```python
contours = measure.find_contours(img_edges, 0.1)
```

```python
fig, ax = plt.subplots()
ax.imshow(img_edges, interpolation='nearest', cmap=plt.cm.gray) # Display the image and plot all contours found
```

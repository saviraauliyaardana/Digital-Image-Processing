```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
cv2.__version__
```

```python
pathj = 'IMG.jpg'
```

```python
pathf = 'haarcascade_frontalface_default.xml'
```

```python
pathe = 'haarcascade_eye.xml'
```

```python
face_cascade = cv2.CascadeClassifier(pathf) # import the trained well face detection function
 
```

```python
eye_cascade = cv2.CascadeClassifier(pathe) # import the trained well eye detection function
```

```python
img = cv2.imread(pathj)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=3, minSize=(32, 32)) # detect  face 
```

```python
j=0
for(x,y,w,h) in faces:
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(100,200,255),10)
    
    face_re_g=gray[y:y+h, x:x+w]
    face_re = img[y:y+h, x:x+w] # draw the face 
    eyes = eye_cascade.detectMultiScale(face_re_g) # eye dection
    i=0
    for(ex,ey,ew,eh) in eyes:
        cv2.rectangle(face_re,(ex,ey),(ex+ew,ey+eh),(200,150,0),10) # draw the eye
        i += 1
        if i == 2:
          break
    j+= 1
    if j == 1:
     break
     
```

```python
# cv2.namedWindow('img', cv2.WINDOW_NORMAL)
plt.figure(figsize=(25,5))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))  # show the result
plt.show ()
```

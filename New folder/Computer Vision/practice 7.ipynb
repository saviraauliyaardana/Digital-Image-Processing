# example of erosion by using skimage

```python
from skimage import io, morphology, color
```

```python
from skimage.morphology import erosion, dilation, opening, closing, white_tophat, disk
```

```python
from skimage.util import img_as_ubyte
```

```python
img=io.imread('lena.png')
```

```python
io.imshow(img)
```

```python
gray = color.rgb2gray(img)
```

```python
io.imshow(gray)
```

```python
img1 = img_as_ubyte(gray)
```

```python
eroded=erosion(img1, disk(2))
```

```python
io.imshow(eroded)
```

# dilation by using skimage, all codes from erosion are the same except the eroded and dilated

```python
dilated = dilation(img1, disk(3))
```

```python
io.imshow(dilated)
```

```python
def plot_comparison(original, filtered, filter_name):

    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(filter_name)
    ax2.axis('off')   #define a convenience function for plotting comparisons
```

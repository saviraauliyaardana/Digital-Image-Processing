# Low level DIP

## start from pillow library

```python
from PIL import Image, ImageFilter
```

```python
img=Image.open('newastronaut.png')
```

```python
display(img)
```

```python
img_blur=img.filter(ImageFilter.GaussianBlur(5))
```

```python
display(img_blur)
```

## now we are going to use skimage to do the DIP

```python
from skimage import io, filters
```

```python
img=io.imread('newastronaut.png')
```

```python
io.imshow(img)
```

```python
img2=filters.gaussian(img,sigma=5)
```

```python
io.imshow(img2)
```

# edge detection from skimage

```python
from skimage import io, color, filters
```

```python
img=io.imread('newastronaut.png')
```

```python
io.imshow(img)
```

```python
img2=color.rgb2gray(img)
```

```python
io.imshow(img2)
```

```python
edge=filters.sobel(img2)
```

```python
io.imshow(edge)
```
